```python
from orm_loader.registry import (
    ModelRegistry,
    ModelDescriptor,
    TableSpec,
    FieldSpec,
    Validator,
    ValidationIssue,
    SeverityLevel,
    ValidationRunner,
    always_on_validators,
)
from pathlib import Path
```

```python
m = ModelRegistry(model_version = '5.4', model_name = 'CDM')
```

```python
field_spec = Path('OMOP_CDMv5.4_Field_Level.csv')
table_spec = Path('OMOP_CDMv5.4_Table_Level.csv')

m.load_table_specs(table_csv=table_spec, field_csv=field_spec)
```

```python
m.discover_models('omop_alchemy.cdm.model')
```

```python
runner = ValidationRunner(
    validators=always_on_validators(),
    fail_fast=False,
)
```

```python
report = runner.run(m)
```

```python
print(report.summary())
```

```python
if not report.is_valid():
    print(report.render_text_report())
```
